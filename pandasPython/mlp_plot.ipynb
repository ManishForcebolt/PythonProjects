```python
import matplotlib.pyplot as plt
import pytrends
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-us')
```

```python


#extract data about keywords
keywords = ['python', 'c', 'c++', 'R', 'Java']
pytrends.build_payload(keywords, timeframe='today 5-y')

# specify and get data
data = pytrends.interest_over_time()
data
```

```python
# type of data
# type(data)  # pandas.core.frame.DataFrame
# plot the data
plt.plot(data)
plt.show()
```

```python

# add title and axes
plt.suptitle('Programming language searches on Google In last 5 years')
# add legends

plt.figure(figsize=(10, 5))
plt.legend(keywords)
plt.xlabel('years')
plt.ylabel('Weekly Searches')

# saving the plot

#plt.savefig('data_plot.png', dpi=150)
plt.plot(data)
plt.show()

```

```python
# analtse data

focus = ['python', 'Java', 'R']

plt.plot(data[focus])
plt.legend(focus)
```

```python
# Extracting country level data 
data2 = pytrends.interest_by_region(resolution='COUNTRY', inc_low_vol=True)
data2 = data2['python'].nlargest(10)
data2
```

```python
# type(data2) # pandas.core.series.Series
# get data to dataframe
data2 = data2.to_frame()
data2
```

```python
# Plot a bar chart
data2.plot(kind='bar')
#data2.plot(kind = 'pi')
plt.suptitle('Top 10 Countries for Python search')
plt.xlabel('Countries')
plt.ylabel('Avg. search in 5 years')
plt.show()
```
